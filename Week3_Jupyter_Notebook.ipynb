```python
# Select all columns with object (string) dtype
string_cols = df.select_dtypes(include='object').columns

# Loop through each string column to detect and remove leading/trailing whitespace
for col in string_cols:
    # Count how many values in this column have extra spaces at start or end
    spaces = df[col].apply(
        lambda x: str(x) != str(x).strip() if pd.notna(x) else False
    ).sum()

    # If any values had whitespace, print the column name and number trimmed
    if spaces > 0:
        print(f"  {col}: {spaces} trimmed")

    # Trim whitespace from all values in the column
    df[col] = df[col].astype(str).str.strip()

# Confirm that trimming is complete
print("All strings trimmed")
```

```python
# Initialize a counter for how many phantom 'nan'/'None' strings we fix
n = 0  

# Loop through all string columns to identify and correct placeholder strings
for col in string_cols:
    # Count cells containing the literal strings 'nan' or 'None'
    c = (df[col] == 'nan').sum() + (df[col] == 'None').sum()

    # If any found, print the column name and count, and update total counter
    if c > 0:
        print(f"  {col}: {c}")  
        n += c

    # Replace the literal strings 'nan' and 'None' with proper NaN values
    df[col] = df[col].replace({'nan': np.nan, 'None': np.nan})

# Print the total number of replacements made
print(f"Converted {n} back to NaN")
```

```python
#  Standardise email to Lowercase

# Print the first 10 email values before standardization
print("BEFORE:")
print(df['email'].head(10))

# Convert all emails to lowercase and remove any leading/trailing whitespace
df['email'] = df['email'].str.lower().str.strip()

# Print the first 10 email values after standardization to verify changes
print("\nAFTER:")
print(df['email'].head(10))
```

```python
# Standardise gender Values 

# Print the counts of all gender values (including NaN) before mapping
print("BEFORE:")
print(df['gender'].value_counts(dropna=False))

# Define a dictionary mapping known variants to standardized labels
gender_map = {
    'male'  : 'Male',
    'm'     : 'Male',
    'female': 'Female',
    'f'     : 'Female',
    'other' : 'Other'
}

# Save a copy of the original column to preserve any unmapped values
orig = df['gender'].copy()

# Normalize to lowercase and map each value using the dictionary; unmapped values become NaN
df['gender'] = df['gender'].str.lower().map(gender_map)

# Restore any original values that were not recognized by the map (preventing accidental NaN)
unmapped = df['gender'].isna() & orig.notna()
if unmapped.sum() > 0:
    df.loc[unmapped, 'gender'] = orig[unmapped]

# Print the counts of all gender values after standardization
print("\nAFTER:")
print(df['gender'].value_counts(dropna=False))
```

```python
# Standardise department Values

print("BEFORE:")
print(df['department'].value_counts(dropna=False))

# Comprehensive mapping covering abbreviations, shorthand, and casing variants.
# Every key is lowercase to match after .str.lower().
dept_map = {
    'eng'                 : 'Engineering',
    'engineering'         : 'Engineering',
    'hr'                  : 'Human Resources',
    'h.r.'                : 'Human Resources',
    'human resources'     : 'Human Resources',
    'sales'               : 'Sales',
    'ops'                 : 'Operations',
    'operations'          : 'Operations',
    'it'                  : 'Information Technology',
    'information technology': 'Information Technology',
    'fin'                 : 'Finance',
    'finance'             : 'Finance'
}

orig = df['department'].copy()
df['department'] = df['department'].str.lower().map(dept_map)

# Restore any values the map did not recognise so we don't silently lose data.
unmapped = df['department'].isna() & orig.notna()
if unmapped.sum() > 0:
    df.loc[unmapped, 'department'] = orig[unmapped]

print("\nAFTER:")
print(df['department'].value_counts(dropna=False))
```

```python
# Standardise job_title Casing

# Job titles in raw HR exports often arrive in inconsistent casing:
# 'senior engineer', 'Senior Engineer', 'SENIOR ENGINEER'.
# Title Case normalises all variants so groupby and reporting work correctly.
print("BEFORE:")
print(df['job_title'].value_counts(dropna=False))

df['job_title'] = df['job_title'].str.title()

print("\nAFTER:")
print(df['job_title'].value_counts(dropna=False))
```

```python
# Standardise currency Values

print("BEFORE:")
print(df['currency'].value_counts(dropna=False))

# Map currency symbols and abbreviations currency codes instead of letters.
curr_map = {
    'r'  : 'ZAR',
    'zar': 'ZAR',
    '£'  : 'GBP',
    'gbp': 'GBP',
    '$'  : 'USD',
    'usd': 'USD'
}

orig = df['currency'].copy()
df['currency'] = df['currency'].str.lower().map(curr_map)

# Restore unrecognised values to avoid data loss.
unmapped = df['currency'].isna() & orig.notna()
if unmapped.sum() > 0:
    df.loc[unmapped, 'currency'] = orig[unmapped]

print("\nAFTER:")
print(df['currency'].value_counts(dropna=False))
```

```python
# Standardise country Values

print("BEFORE:")
print(df['country'].value_counts(dropna=False))

# Create a dictonary map for every known country abbreviation and variant to a full, consistent name.
# convert to full department names  are preferred over codes for readability in HR reporting.
country_map = {
    'u.k.'         : 'United Kingdom',
    'uk'           : 'United Kingdom',
    'united kingdom': 'United Kingdom',
    'sa'           : 'South Africa',
    'za'           : 'South Africa',
    'south africa' : 'South Africa',
    'usa'          : 'United States',
    'united states': 'United States'
}

orig = df['country'].copy()
df['country'] = df['country'].str.lower().map(country_map)

# Restore any values not captured by the map.
unmapped = df['country'].isna() & orig.notna()
if unmapped.sum() > 0:
    df.loc[unmapped, 'country'] = orig[unmapped]

print("\nAFTER:")
print(df['country'].value_counts(dropna=False))
```

```python
# Clean state Column

print("BEFORE:")
print(df['state'].value_counts(dropna=False))

# 'south-africa' is a country name that was incorrectly placed in the state column.
# It carries no meaningful province/state information so we replace it with NaN.
# We will attempt to infer the correct state from the city column in Cell 35.
df['state'] = df['state'].replace('south-africa', np.nan)

# Apply Title Case to ensure consistent formatting across all state names
# (e.g. 'gauteng' -> 'Gauteng', 'western cape' -> 'Western Cape').
df['state'] = df['state'].str.title()

print("\nAFTER:")
print(df['state'].value_counts(dropna=False))
```

```python
#Infer Missing State From City

# South African cities we know the province to it is in with certainty.
# Create a dictionary maps with each city to its correct province/state.
# We use this to recover state values that were missing or incorrectly entered.
city_to_state = {
    'Durban'      : 'KZN',
    'Pretoria'    : 'Gauteng',
    'Polokwane'   : 'Limpopo',
    'Cape Town'   : 'Western Cape',
    'Johannesburg': 'Gauteng'
}

# For each city-state pair, build a boolean mask targeting rows where:
#   - the city matches our lookup key
#   - the state is currently missing (NaN)
# Only fill rows that satisfy BOTH conditions to avoid overwriting valid data.
for city, state in city_to_state.items():
    mask = (df['city'] == city) & (df['state'].isna())
    if mask.sum() > 0:
        df.loc[mask, 'state'] = state
        print(f"  {city} -> {state}: {mask.sum()}")

# Report any rows where state is still missing after the city-based fill.
# These may be international employees where province inference is not possible.
still = df[df['state'].isna() & df['city'].notna()]
if len(still) > 0:
    print(f"Still missing: {len(still)}")
    print(still[['employee_id', 'city', 'country']].to_string())

print(f"\n{df['state'].value_counts(dropna=False)}")
```

```python
# Standardise city Casing

# Apply Title Case to city names for the same reason as names and departments —
# 'cape town', 'Cape Town', and 'CAPE TOWN' must resolve to one canonical form
# so that groupby operations and location-based reporting count them correctly.
df['city'] = df['city'].str.title()

print(df['city'].value_counts(dropna=False))
```

```python
# Standardise and Fill is_active 

# Print the counts of all is_active values (including NaN) before standardization
print("BEFORE:")
print(df['is_active'].value_counts(dropna=False))

# Convert all values to strings, strip whitespace, and uppercase for consistent mapping
df['is_active'] = df['is_active'].astype(str).str.strip().str.upper()

# Map normalized strings to Python booleans; unmapped values become NaN
df['is_active'] = df['is_active'].map({'TRUE': True, 'FALSE': False})

# Count how many NaN values remain before filling
before = df['is_active'].isna().sum()

# Fill NaN values: if exit_date exists, employee has left → False
df.loc[(df['is_active'].isna()) & (df['exit_date'].notna()), 'is_active'] = False

# Fill NaN values: if exit_date is missing, employee likely still active → True
df.loc[(df['is_active'].isna()) & (df['exit_date'].isna()),  'is_active'] = True

# Print how many missing values were filled
print(f"Filled {before - df['is_active'].isna().sum()} missing")

# Print the counts of all is_active values after standardization and filling
print("\nAFTER:")
print(df['is_active'].value_counts(dropna=False))
```

```python
# Fill Missing department Values

# Count how many department values are missing.
m = df['department'].isna().sum()

if m > 0:
    # Fill with the mode (most frequently occurring department).
    # This is a last resort — we use it only when no better signal exists.
    # mode() returns a Series so we index [0] to get the top value as a scalar.
    df['department'] = df['department'].fillna(df['department'].mode()[0])
    print(f"Filled {m} with mode")
else:
    print("No missing departments")
```

```python
# Check whether each email matches a basic pattern (something@something.something)
df['email_valid'] = df['email'].apply(
    # Convert to string, match regex, return True if valid, False otherwise; NaN treated as False
    lambda x: bool(re.match(r'^.+@.+\..+', str(x))) if pd.notna(x) else False
)

# Extract all rows where the email was invalid
inv = df[df['email_valid'] == False]

# Print summary counts of valid and invalid emails
print(f"Valid: {df['email_valid'].sum()} | Invalid: {len(inv)}")

# If there are invalid emails, display their employee details for review
if len(inv) > 0:
    print(inv[['employee_id', 'first_name', 'last_name', 'email']])
```

```python
# Identify rows where salary is negative
neg = df[df['salary'] < 0]

# Print the count of negative salary entries
print(f"Negative salaries: {len(neg)}")

# If there are negative salaries, inspect and correct them
if len(neg) > 0:
    # Show employee_id, first_name, and salary before correction
    print("BEFORE:")
    print(neg[['employee_id', 'first_name', 'salary']])

    # Replace negative salaries with their absolute value, leaving NaN unchanged
    df['salary'] = df['salary'].apply(
        lambda x: abs(x) if pd.notna(x) and x < 0 else x
    )

    # Show the corrected salary values for the affected rows
    print("AFTER:")
    print(df.loc[neg.index, ['employee_id', 'first_name', 'salary']])
```
